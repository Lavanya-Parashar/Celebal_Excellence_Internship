3. Handle missing values

```python
#check missing values
print("Missing values: ")
print(df.isnull().sum())
```

```python
#handling missing values according to specific columns
#for cleaner dataset

df['discount'] = df['discount'].fillna(0)

df['what_customers_said'] = df['what_customers_said'].fillna("No Review")

df['seller_name'] = df['seller_name'].fillna("Unknown Seller")

df['videos'] = df['videos'].fillna("No video")

df['seller_information'] = df['seller_information'].fillna("Not available")

df['variations'] = df['variations'].fillna("No variations")

print("Missing values handled successfully! ")
```

4. Remove Duplicates

```python
#remove duplicates
df.drop_duplicates(inplace=True)
print("Duplicates removed successfully! ")
```

5. Perform basic operations

```python
#displaying products with rating greater than 10
high_rating = df[df['rating']>4]
high_rating.head()
```

```python
#selecting particular columns
selected_columns = df[['title', 'rating', 'final_price']]
selected_columns.head()
```

6. Create Derived Coulmn

```python
# Clean final_price column
df['final_price'] = (
    df['final_price']
    .astype(str)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace('"', '', regex=False)
    .str.strip()
)

# Convert to float
df['final_price'] = pd.to_numeric(df['final_price'])

# Create derived column
# here : final_price acts as price
# ratings_count act as quantity
df['total_amount'] = df['final_price'] * df['ratings_count']

# Check result
df[['final_price', 'ratings_count', 'total_amount']].head()
```

7. Saving the cleaned dataset

```python
df.to_csv('cleaned_tshirts.csv',index=False)
print("Cleaned dataset saved successfully! ")
```
